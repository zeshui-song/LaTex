```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#INPUTS
g = 9.81 #m/s^2
```

### Plot over range of angles

```python
#calculations
q_deg = np.linspace(.1, 80,100)  #angle array in degrees
q_rad = np.radians(q_deg)  #convert to radians
FA_norm = g * np.tan(q_rad)
FC_norm = g /np.cos(q_rad)

# Plot the results
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(q_deg, FA_norm, label=r'Normalized $T_{AB} = g\tan\Theta$', color='blue')
ax.plot(q_deg, FC_norm, label=r'Normalized $T_{BC} = g/\cos\Theta$', color='red')

#Formatting
ax.set_xlabel("Angle Theta (degrees)")
ax.set_ylabel("Normalized Tension")
ax.set_title("Normalized Tensions vs. Angle Theta")
ax.legend()
ax.grid(True)
ax.set_ylim(0, 100)
ax.set_xlim(0, 80)

# Show the plot
plt.show()
```

### Determining maxminum angle

```python
#INPUTS
m = 100  #kg
breaking_load = m * g * 3
```

```python
#Calculations
q_max_rad = np.arccos( (m*g) / (breaking_load) )
q_max_deg = np.degrees(q_max_rad )

#Results
print(f"Maximum angle for mass {m} kg with a breaking load of {breaking_load} N: {q_max_deg:.4g} degrees")
```
