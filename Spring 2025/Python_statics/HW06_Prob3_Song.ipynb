```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Inputs
a = np.linspace(0.001, 0.6,100) #m
q = np.radians(15) #rad
#Equations
P_tip = (10*0.6 - 10*a) / (np.cos(q)*0.9+np.sin(q)*1.2-a*np.sin(q))
mu_s = (P_tip*np.cos(q))/(10-P_tip*np.sin(q))
```

### Plot

```python
# Plot the results
fig, ax1 = plt.subplots()

#Formatting
color = 'tab:blue'
ax1.set_xlabel('Distance a (m)')
ax1.set_ylabel('Applied force P (kN)', color=color)
ax1.plot(a, P_tip, color=color, label='P_tip') 
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend()   
ax1.grid(True) 
ax1.set_ylim(0, 10)
ax1.set_xlim(0, 0.6)

# Secondary axis for moment
ax2 = ax1.twinx()

color = 'tab:red'
ax2.set_ylabel('Coefficient of friction (mu_s,crit)', color=color)
ax2.plot(a, mu_s, color=color, linestyle='--', label='Mu_s,crit')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='lower right', bbox_to_anchor=(1, .8))

# Show the plot
plt.show()
```
