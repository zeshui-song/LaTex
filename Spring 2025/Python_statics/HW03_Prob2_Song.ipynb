```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Inputs
W = 1000 #N
q = 20 #deg
q_rad = np.radians(q)
a = np.linspace(0.1, 69.9,100)
a_rad = np.radians(a)
mu_s = 0.25
```

```python
#Computations
T_lower = (-W*(mu_s*np.cos(q_rad)-np.sin(q_rad))) / (np.sin(a_rad+q_rad)-mu_s*np.cos(a_rad+q_rad))
T_raise = (W*(mu_s*np.cos(q_rad)+np.sin(q_rad))) / (np.sin(a_rad+q_rad)+mu_s*np.cos(a_rad+q_rad))
```

### Plot

```python
# Plot the results
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(a, T_lower, label=r'T_lower', color='orange')
ax.plot(a, T_raise, label=r'T_raise', color='green')

#Formatting
ax.set_xlabel("Angle Alpha (Degrees)")
ax.set_ylabel("Tension (Newtons)")
ax.legend()
ax.grid(True)
ax.set_ylim(0, 1000)
ax.set_xlim(0, 70)

# Show the plot
plt.show()
```
