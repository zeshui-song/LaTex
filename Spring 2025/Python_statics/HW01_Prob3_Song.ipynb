```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#INPUTS
g = 9.81  #m/s^2
FN = 1000  #N
```

### Plot over range of angles

```python
#calculations
q_deg = np.linspace(1, 10,50)  #angle array in degrees
q_rad = np.radians(q_deg)  #convert to radians
T1 = FN / (2* np.sin(q_rad) )

# Plot the results
fig, ax = plt.subplots()
ax.plot(q_deg, T1, label=r'Tension $T_{1} = T_{2} = N/2\sin\Theta$', color='blue')

#Formatting
ax.set_xlabel("Angle Theta (degrees)")
ax.set_ylabel("Tension (N)")
ax.set_title("Tensions vs. Angle Theta (Case B)")
ax.legend()
ax.grid(True)
ax.set_ylim(0, 30000)
ax.set_xlim(1, 10)

# Show the plot
plt.show()
```
