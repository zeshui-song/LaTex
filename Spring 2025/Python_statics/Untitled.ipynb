```python
import numpy as np
import matplotlib.pyplot as plt

# Example range of tensions
tension = np.linspace(0, 1000, 100)

# Dummy formulas for reactions (replace with real equations)
reaction_force = 0.5 * tension + 100   # Just an example formula
reaction_moment = 0.2 * tension**0.5 + 50  # Another example

# Plot
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Tension (N)')
ax1.set_ylabel('Reaction Force (N)', color=color)
ax1.plot(tension, reaction_force, color=color, label='Reaction Force')
ax1.tick_params(axis='y', labelcolor=color)

# Secondary axis for moment
ax2 = ax1.twinx()

color = 'tab:red'
ax2.set_ylabel('Reaction Moment (Nm)', color=color)
ax2.plot(tension, reaction_moment, color=color, linestyle='--', label='Reaction Moment')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Reaction Forces and Moments vs Tension')
fig.tight_layout()
plt.show()
```
