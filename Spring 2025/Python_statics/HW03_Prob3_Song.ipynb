```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Inputs
mu_s = np.linspace(0.1, 1/np.sqrt(3),100)
```

```python
#Computations
alpha = np.arctan(mu_s)
theta = np.arcsin(2*np.sin(alpha)) + alpha
theta_deg = np.degrees(theta)
```

### Plot

```python
# Plot the results
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(mu_s, theta_deg, label=r'theta', color='blue')

#Formatting
ax.set_xlabel("Coefficient of friction")
ax.set_ylabel("Theta (Degrees)")
ax.legend()
ax.grid(True)
ax.set_ylim(0, 120)
ax.set_xlim(0.1, 0.6)

# Show the plot
plt.show()
```
