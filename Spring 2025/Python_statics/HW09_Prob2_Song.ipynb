```python
import numpy as np
```

```python
#Inputs
P = 400 #lb (y)
L_AC = 1/np.sqrt(24**2+10**2)
L_DC = 1/np.sqrt(24**2+7**2)
L_BC = 1/np.sqrt(24**2+7**2)
L_AB = 1/np.sqrt(10**2+7**2)
L_AD = L_AB
```

## Solving for reactions


```python
# Define 6x6 matrix
M = np.array([
    #          Ax     Bx      By     Dx   Dy   Dz
    # 1  
         [      1,     1,      0,     1,   0,   0 ],
    # 2  
         [      0,     0,      1,     0,   1,   0 ],
    # 3  
         [      0,     0,      0,     0,   0,   1 ],
    # 4  
         [      0,     0,    -14,     0,   0,   0 ],
    # 5  
         [      7,    14,      0,     0,   0,   0 ],
    # 6  
         [     10,     0,      0,     0,   0,   0 ],
])


B = np.array([0,P,0,-P*7,0,-P*24])

# Compute
Solution_matrix = np.linalg.solve(M, B)

Ax  = Solution_matrix[0]
Bx  = Solution_matrix[1]
By  = Solution_matrix[2]
Dx  = Solution_matrix[3]
Dy  = Solution_matrix[4]
Dz  = Solution_matrix[5]

print(f'Ax  = {Ax:.4g} lbs')
print(f'Bx  = {Bx:.4g} lbs')
print(f'By  = {By:.4g} lbs')
print(f'Dx  = {Dx:.4g} lbs')
print(f'Dy  = {Dy:.4g} lbs')
print(f'Dz  = {Dz:.4g} lbs')
```

## Analysis at C

```python
F_AC = P / (10*L_AC)
F_DC = -F_AC*(24*L_AC) / (2*24*L_DC)
F_BC = F_DC
```

```python
# Results
print(f'F_AC  = {F_AC:.4g} lbs')
print(f'F_DC  = {F_DC:.4g} lbs')
print(f'F_BC  = {F_BC:.4g} lbs')
```

## Analysis at B

```python
F_BC = -Bx / (24*L_BC)
F_AB = -By / (10*L_AB)
F_BD = -F_BC*(7*L_BC) -F_AB*(7*L_AB)
```

```python
# Results
print(f'F_BC  = {F_BC:.4g} lbs')
print(f'F_AB  = {F_AB:.4g} lbs')
print(f'F_BD  = {F_BD:.4g} lbs')
```

## Analysis at A

```python
F_AC = -Ax / (24*L_AC)
F_AD = ( -F_AC * (10 * L_AC) - F_AB * (10 * L_AB) ) / (10 * L_AD)
```

```python
# Results
print(f'F_AC  = {F_AC:.4g} lbs')
print(f'F_AD  = {F_AD:.4g} lbs')
```

## Sanity check at D

```python
X = Dx + F_DC*(24*L_DC)
Y = Dy + F_AD*(10*L_AD)
Z = Dz + F_BD + F_AD*(7*L_AD) + F_DC*(7*L_DC)

print(f'X  = {X:.4g} lbs')
print(f'Y  = {Y:.4g} lbs')
print(f'Z  = {Z:.4g} lbs')
```
