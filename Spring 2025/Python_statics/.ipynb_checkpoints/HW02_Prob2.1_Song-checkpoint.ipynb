```python
import math
import numpy as np
```

```python
#Inputs
rP = 1000 #kN

#Vector magnitudes
rAD = math.sqrt(3.30**2 + 5.60**2)
rAC = math.sqrt(4.20**2 + 2.40**2 + 5.60**2)
rAB = math.sqrt(4.20**2 + 5.60**2)

#Vector components
P = (0/rP, rP/rP, 0/rP)
AB = (-4.2/rAB, -5.6/rAB, 0/rAB)
AC = (2.4/rAC, -5.6/rAC, 4.2/rAC)
AD = (0/rAD, -5.6/rAD, -3.3/rAD)

#Print vector components
print(f"Unit Vector P:  ({P[0]:.4f}, {P[1]:.4f}, {P[2]:.4f})")
print(f"Unit Vector AB: ({AB[0]:.4f}, {AB[1]:.4f}, {AB[2]:.4f})")
print(f"Unit Vector AC: ({AC[0]:.4f}, {AC[1]:.4f}, {AC[2]:.4f})")
print(f"Unit Vector AD: ({AD[0]:.4f}, {AD[1]:.4f}, {AD[2]:.4f})")
```

```python
#Solving for tensions
A = np.array([
    [AB[0], AC[0], AD[0]],
    [AB[1], AC[1], AD[1]],
    [AB[2], AC[2], AD[2]]
])

B = np.array([0, -rP, 0])

F = np.linalg.solve(A, B)

# Display the result
print(f"F_AB (kN): {F[0]:.4g}")
print(f"F_AC (kN): {F[1]:.4g}")
print(f"F_AD (kN): {F[2]:.4g}")

```
