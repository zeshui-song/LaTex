```python
import numpy as np
```

```python
#Inputs
F = 1 #kN
a = np.radians(60)
d = 2 #m
h = 3 #m
q = np.arctan(d/(2*h))
```

```python
By = (F*np.sin(a)*d - F*np.cos(a)*h)/d
Ay = F*np.sin(a) - By
print(f'By  = {By:.4g} kN')
print(f'Ay  = {Ay:.4g} kN')
```

```python
Hy = (F*np.sin(a)*d) / (d/2)
F_CD = (By - Hy) / np.sin(q)
print(f'Hy  = {Hy:.4g} kN')
print(f'F_CD  = {F_CD:.4g} kN')
```

```python
Bx = (F_CD * ( np.sqrt((d/2)**2 + h**2) / 2 ) - By*(d/2))/h
Hx = Bx - F_CD*np.cos(q)
Ax = F*np.cos(a) - Bx
print(f'Ax  = {Ax:.4g} kN')
print(f'Bx  = {Bx:.4g} kN')
print(f'Hx  = {Hx:.4g} kN')
```

```python
Gx = Hx - F*np.cos(a)
Gy = Hy - F*np.sin(a)
print(f'Gx  = {Gx:.4g} kN')
print(f'Gy  = {Gy:.4g} kN')
```

```python
x = Ax + F_CD*np.cos(q) + Gx
y = Ay + F_CD*np.sin(q) + Gy

print(f'x balanace  = {x:.4g} kN')
print(f'y balance  = {y:.4g} kN')
```

## Results

```python
print(f'Ax  = {Ax:.4g} kN')
print(f'Ay  = {Ay:.4g} kN')
print(f'Bx  = {Bx:.4g} kN')
print(f'By  = {By:.4g} kN')
print(f'Gx  = {Gx:.4g} kN')
print(f'Gy  = {Gy:.4g} kN')
print(f'Hx  = {Hx:.4g} kN')
print(f'Hy  = {Hy:.4g} kN')
print(f'F_CD  = {F_CD:.4g} kN')
```
