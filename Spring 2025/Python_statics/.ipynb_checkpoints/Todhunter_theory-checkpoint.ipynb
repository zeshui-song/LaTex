A schematic of the arrangement is shown that defines the geometry using the 2 angles (wrt vertical) of the cables.  The system could be out of balance, in principle.  That dynamics problem (where all 3 masses accelerate) is actually much more involved than the equilibrium case. A natural way to approach this from a theoretical perspective is to treat the masses as inputs, and the resulting geometry as an output.  That is, set the 3 masses to known values, and determine what angles it makes.  We'll look at it that way in ME103, as it serves as a useful tool for studying 2D particle equilibrium.  However, it is not very practical in that mode.  So in ME104, where quantities are measured, the same apparatus will be looked at where one of the masses is considered known, then the two angles are measured (indirectly) and the equilibrium analysis used to determine the unknown masses.  

There are two pulleys in this set-up (left and right).  A flexible cable that passes over the pulleys is attached to masses L and R at either end.  A short cable that holds M is tied to the cable in between the pulleys.  Point 0 is a "knot" of some kind.  The pulleys will be ASSUMED to be frictionless at this point of the course, which means that the tension in the cable that passes over it is unchanged on either side, and therefore equal to the weight hanging freely (L and R). Friction at the bearing would change that and make the analysis much more difficult.  Experimentally, attempts have been made to minimize that effect, but it cannot be eliminated completely.  That is an important principle related to experimental uncertainty, something to be addressed a bit in ME104. 

Equilibrium is not guaranteed.  If either L or R are too large, they will pull the system to its side (until the other pulley stops it).  If M is greater than the sum of L and R, it will fall until L or R hits its pulley. 

### Free Body Diagrams
Engineering analysis frequently involves isolating some part of the universe and applying basic natural laws (like F=ma) to it.  In thermal/fluids courses, that would be called a "Control Volume."  In mechanics courses, it is a free body diagram (FBD), an indispensible tool in analyis.  It consists of a choice of some "body" to be analyzed, which is pulled out (and made "free").  All forces acting on it are shown acting on the body.  If all the forces act through a single point somewhere, then there is no tendency for the forces to cause rotation, and this is the formal definition of a particle (even if it is large in extent).  The Earth itself commonly behaves like a particle.  Physically, if all the forces act in a plane, then motion is constrained in two directions (not necessarily orthogonal).

An important principle is whether a problem posed is statically determinate or not.  For a 2D particle statics problem (all forces lie in a plane and pass through some common point), there are 2 equations of equilibrium.  Physically, the motion is constrained from moving in two directions.  A problem is statically determinate if there are 2 unknowns.  If there are 3 or more unknowns, then the problem is under-specified and a complete solution is not possible (although a partial solution MIGHT be). It is statically indeterminate. On the other hand, if there is only 1 unknown, the problem is over-specified and there will either be a redundancy or an inconsistency.  In the Todhunter apparatus, there are 3 masses and 2 angles.  Specifying any combination of 3 will result in a statically determinate problem.

### Here's some geometry needed to define the position of the knot:
The origin is placed directly below the left pulley, a distance $y_{left}$.  
The right pulley is a distance $x_{rig}$ to the right,  
at an angle $\theta_{rig}$ from horizontal  
The angles $\alpha$ and $\beta$ are determined from Equilibrium analysis (i.e. Statics).
<div>

</div>


```python
import numpy as np   #Get in the habit of importing numpy like this for numerical work
import matplotlib.pyplot as plt  #and this to make plots.
```

```python
## INPUT PARAMETERS
# L is mass on left, R is mass on right, M is mass in middle
# Default Mass Values
L = 93.6  #kg
R = 98.33  #kg
M = 100 #kg

# RUN DIFFERENT CASES BY UN-COMMENTING THESE LINES
#L = 50  #kg
#R = 51  #kg
#M = 100 #kg

# Fixed Experimental Rig Geometry
x_rig = 1.2   #m, horizontal distance between pulleys
y_left = 1   #m, vertical distance from bottom to left pulley
y_right = 1.3  #m, vertical distance from bottom to right pulley

```

```python
#RESULTS
c = (x_rig**2 + (y_right - y_left)**2)**0.5
alpha_rig = np.arctan(x_rig/(y_right - y_left))   #rad, angle between pulleys
alpha_arg = (-R**2+L**2+M**2)/(2*L*M)  #rad, argument for angle between left leg and vertical
beta_arg = (-L**2+R**2+M**2)/(2*R*M)   #rad, argument angle between right leg and vertical
if abs(alpha_arg) < 1. and abs(beta_arg) < 1:
    alpha = np.arccos(alpha_arg)
    beta = np.arccos(beta_arg)
    print(f'alpha = {alpha*180/np.pi:.4g} deg, beta = {beta*180/np.pi:.4g} deg')    
    if (beta < alpha_rig) and (alpha < np.pi/2 + alpha_rig):
        a = c*np.sin(alpha_rig + alpha)/np.sin(alpha+beta)  #m, distance from left pulley to knot
        x_knot = a*np.sin(alpha)   #m, knot location RELATIVE TO LEFT PULLEY
        y_knot = y_left-a*np.cos(alpha)  #m, knot location RELATIVE TO LEFT PULLEY
        #print(f'a ={a:.4g}m, b = {b:.4g}m')
        print(f'Knot is at x = {x_knot:.4g} m, y = {(y_knot):.4g} m')
    else:
        print(f'Rig angle between pulleys too large for this case')
else:
    print('Equilibrium is Not Possible for this case')    

```

```python
%whos
#It can be a good debugging tool to add a "who" cell to see what you're getting, check numbers, variable types, etc.
```

```python
#Plot Result:
fig, ax = plt.subplots()
ax.plot([0,0,x_knot],[0,y_left, y_knot],'ro-',markersize=10)
ax.plot([x_rig, x_rig,x_knot],[0,y_right, y_knot],'bo-', markersize = 10)
ax.plot([x_knot,x_knot],[y_knot, y_knot - .4],'go-',markersize = 12)
ax.set_xlabel('x position (m)')
ax.set_ylabel('y position (m)')
ax.set_title(f'With L = {L:.3g}, M = {M:.3g}, R = {R:.3g} kg')
ax.axis('equal')
plt.show()
```
