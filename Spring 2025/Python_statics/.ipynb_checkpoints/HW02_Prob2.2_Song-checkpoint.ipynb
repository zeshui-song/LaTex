```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Inputs
W = 100 #N

#Vector magnitudes
rF = math.sqrt(1)
rAB = math.sqrt(1**2 + 1**2 + 1**2)
rAC = math.sqrt(1**2 + 1**2 + 2**2)

#Vector components
AB = (-1/rAB, -1/rAB, 1/rAB)
AC = (-1/rAC, 1/rAC, 2/rAC)

#Print vector components
print(f"Unit Vector F:  (cos(q_rad), 0, sin(q_rad))")
print(f"Unit Vector AB: ({AB[0]:.4f}, {AB[1]:.4f}, {AB[2]:.4f})")
print(f"Unit Vector AC: ({AC[0]:.4f}, {AC[1]:.4f}, {AC[2]:.4f})")
```

```python
#Solving for F
def solve(theta):
    A = np.array([
        [np.cos(theta), AB[0], AC[0]],
        [0, AB[1], AC[1]],
        [np.sin(theta), AB[2], AC[2]]
    ])

    B = np.array([0, 0, W])
    X = np.linalg.solve(A, B)
    return X

# Display the result
print(f"At θ = 90°:")
print(f"F = {solve(np.pi/2)[0]:.4g} N")
print(f"T_AB = {solve(np.pi/2)[1]:.4g}")
print(f"T_AC = {solve(np.pi/2)[2]:.4g}")

#Sanity check: at 90 degrees, F is vertical and supporting the entirety of W. At this angle, T_AB and T_AC should be zero and F should be 100 N.
```

### Plot over range of angles

```python
q_deg = np.linspace(.1, 90,100)  #angle array in degrees
q_rad = np.radians(q_deg)  #convert to radians
F = [solve(theta)[0] for theta in q_rad] #F as a function of theta
T_AB = [solve(theta)[1] for theta in q_rad] #T_AB as a function of theta
T_AC = [solve(theta)[2] for theta in q_rad] #T_AC as a function of theta

# Plot the results
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(q_deg, F, label=r'F', color='blue')
ax.plot(q_deg, T_AB, label=r'T_AB', color='red')
ax.plot(q_deg, T_AC, label=r'T_AC', color='green')

#Formatting
ax.set_xlabel("Angle (degrees)")
ax.set_ylabel("Force (newtons)")
ax.legend()
ax.grid(True)
ax.set_ylim(0, 100)
ax.set_xlim(0, 90)

# Show the plot
plt.show()
```

### Minimizing Force F

```python
idx = np.argmin(F) #finds index where F is minimum
print(f"Minimum F = {F[idx]:.4g} N when angle theta = {idx:.4g} degrees") #Thanks to Prof.Sidebotham for the help on this function
```
