```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Inputs
P = 1000 #N
L = 25 #in
```

```python
#Computations
x = np.linspace(1, 14,100)

z = np.sqrt(225-x**2)
R_A = np.sqrt( (20 * P/x)**2 + (P**2 * (225-x**2)/x**2) )
R_B = np.sqrt( P**2 + 400*P**2/x**2 )
T = 25*P/x
Q = P*np.sqrt( 225-x**2 )/x
MA = Q/P
```

### Plot of z

```python
# Plot the results
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, z, label=r'Z', color='orange')

#Formatting
ax.set_xlabel("x Distance (Inches)")
ax.set_ylabel("z Distance (Inches)")
ax.legend()
ax.grid(True)
ax.set_ylim(0, 15)
ax.set_xlim(1, 14)

# Show the plot
plt.show()
```

### Plot of Reaction Forces

```python
# Plot the results
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, R_A, label=r'R_A', color='blue')
ax.plot(x, R_B, label=r'R_B', color='red')

#Formatting
ax.set_xlabel("x Distance (Inches)")
ax.set_ylabel("Force (newtons)")
ax.legend()
ax.grid(True)
ax.set_ylim(0, 25000)
ax.set_xlim(1, 14)

# Show the plot
plt.show()
```

### Plot of Mechanical Advantage

```python
# Plot the results
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, MA, label=r'Mechanical Advantage Q/P', color='green')

#Formatting
ax.set_xlabel("x Distance (Inches)")
ax.set_ylabel("Mechanical Advantage")
ax.legend()
ax.grid(True)
ax.set_ylim(0, 15)
ax.set_xlim(1, 14)

# Show the plot
plt.show()
```
