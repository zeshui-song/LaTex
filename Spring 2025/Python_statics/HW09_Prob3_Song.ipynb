```python
import numpy as np
```

```python
#Inputs
P = 4
L = 5
```

```python
F_35 = -(0.481*P) / (np.sqrt(3)/2)
F_19 = ((0.481*P)*(2.5*L) - P*L*(3*np.sqrt(3)/2) + P*L*(2*np.sqrt(3)/2)) / (L*np.sqrt(3)/2)
F_21 = -F_19 - 0.5*F_35

print(f'F_35  = {F_35/P:.4g} P')
print(f'F_19  = {F_19/P:.4g} P')
print(f'F_21  = {F_21/P:.4g} P')
```

```python
F_5 = (-0.518*P*2*L + 0.4*P*L) / (L*np.sqrt(3)/2)
F_33 = (-0.519*P + 0.4*P - F_5*(np.sqrt(3)/2)) / (np.sqrt(3)/2)
F_2 = -F_33*(1/2) - F_5*(1/2)

print(f'F_5  = {F_5/P:.4g} P')
print(f'F_33  = {F_33/P:.4g} P')
print(f'F_2  = {F_2/P:.4g} P')
```
