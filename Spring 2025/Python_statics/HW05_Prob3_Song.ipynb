```python
import math
import numpy as np
```

```python
#Inputs
F = 120 #lbs
q = 30 *np.pi/180 #rad

#Vector components
NA = (1,0,24*np.sin(q))
NB = (np.sin(q),-np.cos(q),16)
NC = (-np.sin(q),np.cos(q),-8)
```

```python
#Solving for tensions
A = np.array([
    [NA[0], NB[0], NC[0]],
    [NA[1], NB[1], NC[1]],
    [NA[2], NB[2], NC[2]]
])

B = np.array([0, F, 0])

N = np.linalg.solve(A, B)

# Display the result
print(f"NA (lbs): {N[0]:.4g}")
print(f"NB (lbs): {N[1]:.4g}")
print(f"NC (lbs): {N[2]:.4g}")

```
