```python
import numpy as np
```

## Solving for Tension

```python
#Inputs
lambda_AD = np.array([1/np.sqrt(1+0.75**2), 0, -0.75/np.sqrt(1+0.75**2)])
r_AB = np.array([0.5, 0, 0])
lambda_BG_BH = np.array([-0.5/np.sqrt(0.5**2 + 0.925**2 + (0.75-0.35)**2) + (0.875-0.5)/np.sqrt((0.875-0.5)**2 + 0.75**2 + 0.75**2), 
                         0.925/np.sqrt(0.5**2 + 0.925**2 + (0.75-0.35)**2) + 0.75/np.sqrt((0.875-0.5)**2 + 0.75**2 + 0.75**2), 
                         -(0.75-0.35)/np.sqrt(0.5**2 + 0.925**2 + (0.75-0.35)**2) + -0.75/np.sqrt((0.875-0.5)**2 + 0.75**2 + 0.75**2)])

r_AC = np.array([1, 0, 0])
P = np.array([0, -100, 0])
```

```python
#Define matrices
C1_matrix = np.vstack([lambda_AD, r_AB, lambda_BG_BH])
D2_matrix = np.vstack([lambda_AD, r_AC, P])

# Compute determinants
C1 = np.linalg.det(C1_matrix)
D2 = np.linalg.det(D2_matrix)

# Solve for T
T = -D2 / C1
print(f'T = {T:.4g} kN')
```

## Solving for Thrust from Tension and P

```python
lambda_BG = np.array([-0.5/np.sqrt(0.5**2 + 0.925**2 + (0.75-0.35)**2), 
                      0.925/np.sqrt(0.5**2 + 0.925**2 + (0.75-0.35)**2), 
                      -(0.75-0.35)/np.sqrt(0.5**2 + 0.925**2 + (0.75-0.35)**2)])
lambda_BH = np.array([(0.875-0.5)/np.sqrt((0.875-0.5)**2 + 0.75**2 + 0.75**2), 
                      0.75/np.sqrt((0.875-0.5)**2 + 0.75**2 + 0.75**2), 
                      -0.75/np.sqrt((0.875-0.5)**2 + 0.75**2 + 0.75**2)])

Thrust = np.dot(lambda_AD, T*lambda_BG) + np.dot(lambda_AD, T*lambda_BH)
print(f'Thrust = {Thrust:.4g} kN')
```
