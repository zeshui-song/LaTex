```python
import numpy as np

# Coefficient matrix (ab values)
A = np.array([
    [0.3824, 0.101],
    [0.001,  0.8754]
])

# Measured absorbances at 435 nm and 618 nm
b = np.array([0.322, 0.107])

# Solve the linear system A * x = b
concentrations = np.linalg.solve(A, b)

# Extract concentrations
HIn_conc = concentrations[0]
In_conc = concentrations[1]

# Print the results
print(f"[HIn] = {HIn_conc:.3f}")
print(f"[In⁻] = {In_conc:.3f}")
```
