```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#INPUTS
W = 1000 #N, applied load 
q_deg = 1    #degrees
```

```python
#Calculations
q_rad = q_deg*np.pi/180
print(f'Angle in radians = {q_rad:.4g}')
FA = -W/2/np.sin(q_rad)
print(f'FA = FC = {FA:.6f}')
```

## Plot over range of angles

```python
q = np.linspace(.1,10,5)*np.pi/180   #angle array in radians
FA_range = W/2/np.sin(q)
```

```python
fig,ax = plt.subplots()
ax.plot(q,FA_range)
```
