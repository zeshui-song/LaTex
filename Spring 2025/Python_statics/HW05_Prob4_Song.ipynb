```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Inputs
d = np.linspace(0.001, 0.8999,100)

#Equations
A = 90/(0.9-d)
B_Y = 300 - 90/(0.9-d)
```

### Plot

```python
# Plot the results
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(d, A, label=r'A', color='red')
ax.plot(d, B_Y, label=r'B_y', color='blue')

#Formatting
ax.set_xlabel("d (Meters)")
ax.set_ylabel("Reaction Force (N)")
ax.legend()
ax.grid(True)
ax.set_ylim(-100, 400)
ax.set_xlim(0, 0.7)

# Show the plot
plt.show()
```
