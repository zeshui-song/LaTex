```python
import numpy as np
```

```python
#Inputs
P = 100 #kN
Q = 200 #kN
L = 4
```

```python
F_12 = (3*P*L) / (L*np.sqrt(3)/2)
F_11 = -P / (np.sqrt(3)/2)
F_10 = -F_12 - F_11*(1/2)

print(f'F_12  = {F_12:.4g} kN')
print(f'F_11  = {F_11:.4g} kN')
print(f'F_10  = {F_10:.4g} kN')
```

```python
# Define 2x2 matrix
M = np.array([
    #         F_13               F_14   
    # 1  
         [    1/2,               -1/2             ],
    # 2  
         [    -np.sqrt(3)/2,     -np.sqrt(3)/2    ],
])


B = np.array([-F_12,0])

# Compute
Solution_matrix = np.linalg.solve(M, B)

F_13  = Solution_matrix[0]
F_14  = Solution_matrix[1]

print(f'F_13  = {F_13:.4g} kN')
print(f'F_14  = {F_14:.4g} kN')
```
