```python
import numpy as np
```

```python
#Inputs
P = 150 #lb
a = np.arctan(1.5/3)
```

```python
D = (-3*P) / (np.cos(a)*1.5 +np.sin(a)*3)
Cx = D*np.cos(a)
Cy = P - D*np.sin(a)
print(f'D  = {D:.4g} lbs')
print(f'Cx  = {Cx:.4g} lbs')
print(f'Cy  = {Cy:.4g} lbs')
```

```python
x = -D*np.cos(a)*1.5 + D*np.sin(a)*6 + Cx*3 +Cy*3
print(f'Moment at A  = {x:.4g} lb-ft')
```
