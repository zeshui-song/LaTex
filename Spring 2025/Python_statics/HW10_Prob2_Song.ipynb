```python
import numpy as np
```

```python
#Inputs
P = 150 #lb
E = 100 #lb
```

```python
Cx = -P
Bx = -Cx
print(f'Cx  = {Cx:.4g} lbs')
print(f'Bx  = {Bx:.4g} lbs')
```

```python
By = (1.5*E - Bx*1.5)/3
Cy = P - By
print(f'By  = {By:.4g} lbs')
print(f'Cy  = {Cy:.4g} lbs')
```

```python
Dx = Bx
Dy = By - E
print(f'Dx  = {Dx:.4g} lbs')
print(f'Dy  = {Dy:.4g} lbs')
```

## Results

```python
print(f'Cx  = {Cx:.4g} lbs')
print(f'Cy  = {Cy:.4g} lbs')
print(f'Bx  = {Bx:.4g} lbs')
print(f'By  = {By:.4g} lbs')
print(f'Dx  = {Dx:.4g} lbs')
print(f'Dy  = {Dy:.4g} lbs')
```

```python
x = 4.5*E + 3*Cy + 3*Cx + 6*Dy + 1.5*Dx
print(f'Moments about A  = {x:.4g} lbs')
```
