```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Inputs
T = np.linspace(0.001, 3006.3,100)

#Equations
Rx = 450 - T*(250/np.sqrt(600**2+250**2))
Ry = 750 + T*(600/np.sqrt(600**2+250**2))
Mc = -555000 + T*(369.2)
```

### Plot

```python
# Plot the results
fig, ax1 = plt.subplots()

#Formatting
color = 'tab:blue'
ax1.set_xlabel('Tension (N)')
ax1.set_ylabel('Reaction Force (N)', color=color)
ax1.plot(T, Rx, color=color, label='Rx')
ax1.plot(T, Ry, color='green', label='Ry')  
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend()   
ax1.grid(True) 
ax1.set_ylim(-1000, 2000)
ax1.set_xlim(0, 3006.3)

# Secondary axis for moment
ax2 = ax1.twinx()

color = 'tab:red'
ax2.set_ylabel('Reaction Moment (Nmm)', color=color)
ax2.plot(T, Mc, color=color, linestyle='--', label='Mc')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='lower right', bbox_to_anchor=(1, .65))

# Show the plot
plt.title('Reaction Forces and Moments vs Tension')
plt.show()
```
