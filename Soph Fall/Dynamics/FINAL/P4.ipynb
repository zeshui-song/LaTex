```python
import numpy as np
# --------------------------------------------------
# Constants
# --------------------------------------------------
m1 = 0.064 #rod BAC kg
m2 = 0.043 #rod BD kg
I1 = 4.8e-4 #rod BAC kg*m^2
I2 = 1.4333e-4 #rod BD kg*m^2
g = 9.81 #m/s^2

# Coefficient matrix
A = np.array([
    [0,     0,     1,     0,     0,        -0.01*m2],
    [0,     0,     0,     1,     0.08*m2,   0],
    [0,     0,     0.01,  0,     0,        -I2],
    [1,     0,    -1,     0,     0,         0],
    [0,     1,     0,    -1,    -0.07*m1,   0],
    [0,    -0.07,  0,     0.15, -I1,     0]
])

# Righ hand side
b = np.array([
    0.32*m2,
    m2*g,
    0.0,
    -0.28*m1,
    m1*g,
    0.0
])

# Solve the system
x = np.linalg.solve(A, b)

# Unpack results
Ax, Ay, Bx, By, alpha1, alpha2 = x

# Print results
print(f"A_x = {Ax:.4g} N")
print(f"A_y = {Ay:.4g} N")
print(f"B_x = {Bx:.4g} N")
print(f"B_y = {By:.4g} N")
print(f"alpha_1 = {alpha1:.4g} rad/s^2")
print(f"alpha_2 = {alpha2:.4g} rad/s^2")
```
